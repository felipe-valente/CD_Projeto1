**Pergunta:** Pases que passam a fazer parte do MERCOSUL tem um maior crescimento anual per capta do PIB?

**Hiptese inicial:**  A partir do momento que o pas passa a fazer parte do MERCOSUL, o seu PIB per capta tende a aumentar em relao ao passado.

**Mecanismo:** O MERCOSUL prev: livre circulao de bens e servios atravs dos cortes de direitos alfandegrios e impostos sobre os tais, compromisso para harmonizar as leis dos Estados participantes visando fortalecer o processo de integrao, alm de outros acordos assinados no Tratado de Assuno. Por isso o comercio entre os pases aumentaria e assim aquecendo o mercado sul-americano, aumentando o PIB e consequentemente o PIB per capita.
