**Pergunta:** Países que passam a fazer parte do MERCOSUL tem um maior crescimento anual per capta do PIB?

**Hipótese inicial:**  A partir do momento que o país passa a fazer parte do MERCOSUL, o seu PIB per capta tende a aumentar em relação ao passado.

**Mecanismo:** O MERCOSUL prevê: livre circulação de bens e serviços através dos cortes de direitos alfandegários e impostos sobre os tais, compromisso para harmonizar as leis dos Estados participantes visando fortalecer o processo de integração, além de outros acordos assinados no Tratado de Assunção. Por isso o comercio entre os países aumentaria e assim aquecendo o mercado sul-americano, aumentando o PIB e consequentemente o PIB per capita.
