# **Projeto 1**

# **Pergunta** 
Países que passam a fazer parte do MERCOSUL como membros plenos têm um maior crescimento anual per capita do PIB?

# **Hipótese inicial** 
A partir do momento que o país assina o Tratado de Assunção e passa a fazer parte do MERCOSUL, o seu PIB per capta tende a aumentar em relação ao passado.

# **Mecanismo** 
O MERCOSUL prevê: livre circulação de bens e serviços através dos cortes de direitos alfandegários e impostos sobre os tais, como diz o Artigo 1 do Tratado de Assunção, compromisso para harmonizar as leis dos Estados participantes visando fortalecer o processo de integração, como diz o Artigo 2 do Tratado de Assunção, além de outros acordos assinados. Por isso o livre comercio entre os países aumentaria e assim aquecendo o mercado sul-americano, aumentando o PIB e consequentemente o PIB per capita.

**Artigo 1 do Tratado de Assunção**

> Os Estados Partes decidem constituir um Mercado Comum, que deverá estar estabelecido a 31 de
dezembro de 1994, e que se denominará "Mercado Comum do Sul" (MERCOSUL).      
Este Mercado Comum implica:                                                  
A livre circular de bens serviços e fatores produtivos entre os países entre outros, da eliminação dos
direitos alfandegários restrições não tarifárias à circulação de mercado de qualquer outra medida de
efeito equivalente;                                                         
                                                                            O estabelecimento de uma tarifa externa comum e a adoção de uma política comercial comum em
relação a terceiros Estados ou agrupamentos de Estados e a coordenação de posições me foros
econômico-comerciais regionais e internacionais;                                
A coordenação de políticas macroeconômicas e setoriais entre os Estados Partes - de comércio
exterior, agrícola, industrial, fiscal, monetária, cambial e de capitais, de serviços, alfandegária, de
transportes e comunicações e outras que se acordem -, a gim de assegurar condições adequadas de
concorrência entre os Estados Partes; e                                       
O compromisso dos Estados Partes de harmonizar suas legislações, nas áreas pertinentes, para lograr
o fortalecimento do processo de integração.

**Artigo 2 do Tratado de Assunção**

> O Mercado Comum estará fundado na reciprocidade de direitos e obrigações entre os Estados Partes.

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import os
from numpy import arange
from numpy import percentile
import numpy as np
from IPython.display import clear_output
#repositorio git: https://github.com/felipe-valente/CD_Projeto1.git
```

```python
dados = pd.read_excel('GDP_per_capta.xlsx')
dados
```

```python
data = dados.set_index('GDP per capita growth (annual %)')
a = data.transpose()
mundo = a.loc['1971':'2010']
```

```python
mercosul1 = data.loc[['Brazil','Argentina','Paraguay','Uruguay']]
mercosul2 = mercosul1.transpose()
mercosul3 = mercosul2[mercosul2.index < '2011']
mercosul4 = mercosul3[mercosul3.index > '1970']
mercosul4.insert(4, 'Média MERCOSUL', (mercosul4.transpose().mean()))
mercosul4.insert(5, 'Média mundo', (mundo.transpose().mean()))
pre_mercosul = mercosul4[mercosul4.index < '1991']
pre_mercosul
pos_mercosul = mercosul4[mercosul4.index > '1990']
pos_mercosul
```

```python
mundo_pre_mercosul = mundo[mundo.index < '1991']
mundo_pos_mercosul = mundo[mundo.index > '1990']
mundo_pre_mercosul
mundo_pos_mercosul
```

# **Contexto da análise**

Foi observado o comportamento dos 4 países que assinaram o Tratado de Assunção (Brasil, Argentina, Paraguai e Uruguai) e o comportamento do resto do mundo em comparação com cada país do MERCOSUL. O período de observação foi de 1971 (20 anos antes do Tratado de Assunção) a 2010 (20 anos depois do Tratado de Assunção).

# Brasil

```python
pre_mercosul['Brazil'].describe()
pos_mercosul['Brazil'].describe()

```

```python
plt.boxplot([pre_mercosul['Brazil'], pre_mercosul['Média mundo']], labels = ('Brasil pré MERCOSUL', 'Média do mundo pré MERCOSUL'))
plt.grid(True)
clear_output()
plt.show()
```

```python
plt.boxplot([pos_mercosul['Brazil'], pos_mercosul['Média mundo']], labels = ('Brasil pós MERCOSUL', 'Média do mundo pós MERCOSUL'))
plt.grid(True)
clear_output()
plt.show()
```

Observando os dados do Brasil através do boxplot e do describe percebe-se que a média do Brasil caiu depois que ele entrou no MERCOSUL, o que em um primeiro momento pode indicar que não foi beneficial entrar no bloco. Após o Brasil assinar o Tratado de Assunção o seu comportamento parece começar a se assemelhar mais com o o do resto do mundo por ter diminuído o desvio padrão, o que sugere que o país se tornou mais estável, e o terceiro quartil estar quase nivelado com os dados mundiais, como se pode notar no segundo boxplot do Brasil, porém analisar o comportamento do país que varia ao longo do tempo por meio de um boxplot, não é muito adequado, além de ser um pouco inconclusivo. Portanto foi plotado um gráfico temporal do crescimento anual do PIB per capita do Brasil antes e depois de entrar no MERCOSUL em conjunto com o do mundo (gráfico 1).


```python
anos1 = np.arange(1971,1991,1)
anos2 = np.arange(1991,2011,1)
anos3 = np.arange(1971,2011,1)
plt.plot(anos3, mercosul4['Média mundo'], 'k', label = 'PIB per capita do mundo')
plt.plot(anos1, pre_mercosul['Brazil'], 'g', label = 'Brasil pré MERCOSUL')
plt.plot(anos2, pos_mercosul['Brazil'], 'y', label = 'Brasil pós MERCOSUL')
plt.legend()
plt.xlabel('Tempo (anos)')
plt.ylabel('PIB per capita (anual %)')
plt.title('Gráfico 1')
plt.grid(True)
clear_output()
plt.show()
```

Analisando o Brasil ao longo do tempo percebe-se que de 1971 até 1990 havia muita flutuação do crescimento do PIB per capita anual, valores que comprados com o do resto mundo são muito distantes, sendo um deles muito notável de 1980 para 1981, justamente quando houve a segunda crise do petróleo, que afetou muitos países em desenvolvimento. Quando observados os anos seguintes a partir de 1991, observa-se um Brasil com dados menos flutuantes, mais estáveis e que se assemelham mais aos dados mundiais, com valores mais próximos e com um formato mais semelhante, o que mostra uma estabilização em relação aos dados mundiais e uma melhora em relação aos dados do passado, o que sugere que foi beneficial para o Brasil fazer parte do MERCOSUL.

```python
pre_mercosul['Argentina'].describe()
pos_mercosul['Argentina'].describe()
```

```python
plt.boxplot([pre_mercosul['Argentina'], pre_mercosul['Média mundo']], labels = ('Argentina pré MERCOSUL', 'Média do mundo pré MERCOSUL'))
plt.grid(True)
clear_output()
plt.show()
```

```python
plt.boxplot([pos_mercosul['Argentina'], pos_mercosul['Média mundo']], labels = ('Argentina pós MERCOSUL', 'Média do mundo pós MERCOSUL'))
plt.grid(True)
clear_output()
plt.show()
```

Na Argentina aconteceu o contrário do que no Brasil, a média aumentou significativamente, com um crescimento de 4,22%, o que pode indicar que ao entrar no MERCOSUL o país tem um crescimento no PIB per capita anual, e se comparado com o mundo a média da Argentina sempre esteve próxima. Por outro lado, o desvio padrão aumentou 0,98%, ele que já era muito maior do que o do mundo, mesmo que não seja tão significativo o país já tinha um desvio padrão muito alto, um aumento por mais mínimo que seja sugere um aumento na já alta instabilidade do país. Mas assim como no Brasil um boxplot não é o suficiente para entender o comportamento do crescimento do PIB per capita ao ano, então foi plotado um gráfico no tempo (Gráfico 2).

```python
anos1 = np.arange(1971,1991,1)
anos2 = np.arange(1991,2011,1)
anos3 = np.arange(1971,2011,1)
plt.plot(anos3, mercosul4['Média mundo'], 'k', label = 'PIB per capita do mundo')
plt.plot(anos1, pre_mercosul['Argentina'], 'b', label = 'Argentina pré MERCOSUL')
plt.plot(anos2, pos_mercosul['Argentina'], 'y', label = 'Argentina pós MERCOSUL')
plt.legend()
plt.xlabel('Tempo (anos)')
plt.ylabel('PIB per capita (anual %)')
plt.title('Gráfico 2')
plt.grid(True)
clear_output()
plt.show()
```

Analisando a Argentina de 1971 a 1990 nota-se também uma alta flutuação dos dados, mudando de maneira brusca anualmente de forma que não há como comparar com o resto do mundo pela alta mudança ano a ano. Assim como no Brasil também houve uma queda notável de 1980 para 1981, por conta da segunda crise do petróleo. Nos anos seguintes, de 1991 a 2010 tem o valor mais baixo de crescimento de PIB per capita anual do período analisado e isso se dá por conta da crise da Argentina que se deu entre 2001 e 2002 em que o governo não possuía fundos para manter a paridade fixa ante o dólar, apesar disso apresentou uma melhora significativa no ano seguinte e os dados começaram a se assemelhar mais ainda com os dados mundiais, o que pode se dar ao fato de fazer parte do MERCOSUL.

No Paraguai a média passou de 3,08% nos primeiros 20 anos analisados para 0,82% nos 20 anos conseguintes, algo que inicialmente sugere que não foi beneficial entrar no MERCOSUL, e olhando para o desvio padrão, pode-se notar que houve uma mudança mínima, pois era antes de 4,61% e passou a ser de 4,01%, o que indica que o país adquiriu mais estabilidade ao entrar para o bloco. Analisando o Paraguai de 1971 a 1990 observa-se que ele vinha com um bom crescimento até a segunda crise do petróleo, mas se recuperou bem e se aproximou dos dados mundiais pouco antes de entrar para o MERCOSUL (Gráfico 3). Pouco tempo depois de entrar no bloco o Paraguai teve uma queda do crescimento ficando bem abaixo dos dados mundiais, isso pode ter acontecido por conta do Efeito Tequila, efeito que foi denominado após a crise do peso mexicano, ou até mesmo da própria crise na Argentina. Não tão rápido quanto os argentinos, os paraguaios se recuperaram e por volta de 2003 já estava com números próximos dos dados mundiais. Observando o seu comportamento em geral após aderir ao MERCOSUL o país se mostrou um pouco mais estável e mostrou que apesar de ter sido afetado pela crise imobiliária de 2008 conseguiu se recuperar muito bem de 2009 para 2010 e de ter sofrido baixas de 1996 até 2000 o país parece ter se beneficiado do bloco por ter uma recuperação consistente de crises e por ficar próximo do crescimento do mundo na maior parte do tempo após entrar no bloco.

# **Conclusão**

O Brasil foi o país que sofreu menos com as crises no que se refere ao crescimento anual do PIB per capita, e se mostrou o país mais estável dos analisados, mantendo-se sempre próximo dos dados mundiais após entrar no MERCOSUL, a Argentina sofreu com crises mas demonstrou uma rápida recuperação assim como o Uruguai, o Paraguai se manteve razoavelmente próximo aos dados mundiais e se mostrou mais estável após ter assinado o Tratado de Assunção, logo conclui-se que os países que assinaram o tratado ou passam a ter maior estabilidade, ou se recuperam melhor de crises do que quando não faziam parte do bloco, assim entende-se que fazer parte do MERCOSUL contribui positivamente para o país, não necessariamente aumentando o crescimento anual do PIB per capita, mas fazendo com que ele seja mais estável ou mais próximo dos dados mundiais.
